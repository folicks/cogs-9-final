```python
import kagglehub
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Download latest version
path = kagglehub.dataset_download("manishabhatt22/marketing-campaign-performance-dataset")

print("Path to dataset files:", path)
```

```python


print("Files in dataset directory:", os.listdir(path))

csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
print("CSV files found:", csv_files)

if csv_files:
    csv_path = os.path.join(path, csv_files[0])
    df = pd.read_csv(csv_path)
    print("Variables in the dataset (columns):")
    print(df.columns.tolist())
else:
    print("No CSV files found in the dataset directory.")
```

```python
# prompt: can you provide me few pie charts on the columns of df Customer_Segment, Language, Location, Duration. then may you make it tabulated like into the size of about 500 * 500 output changing the approriate matplot dimensions

import matplotlib.pyplot as plt

# Assuming df is already loaded as in the previous code

cols_to_plot = ['Customer_Segment', 'Language', 'Location', 'Duration']

# Set figure size (adjust as needed)
plt.figure(figsize=(6, 6))  # Adjust width and height for 500x500 output (approximate)


for i, col in enumerate(cols_to_plot):
    if col in df.columns:
        plt.subplot(2, 2, i + 1)  # Create subplots for each column

        try:
            df[col].value_counts().plot.pie(autopct='%1.1f%%')
            plt.title(col)

        except TypeError:  # Handle potential errors if the column data type is not suitable
          print(f"Could not create a pie chart for {col}.")
          plt.title(f"Error for {col}")


plt.tight_layout()
plt.show()
```

```python
# prompt: can you get this a viz to put roi on y and channel used x specificlally make a tabulate of 6 boxplots for the channels present in the dataset

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and it contains 'ROI' and 'Channel_Used' columns.

if 'ROI' in df.columns and 'Channel_Used' in df.columns:
    plt.figure(figsize=(12, 6))  # Adjust figure size as needed

    # Create the boxplot
    sns.boxplot(x='Channel_Used', y='ROI', data=df)

    # Customize the plot
    plt.xlabel('Channel Used', fontsize=12)
    plt.ylabel('ROI', fontsize=12)
    plt.title('ROI Distribution by Channel', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels
    plt.tight_layout()  # Adjust layout to prevent labels from overlapping
    plt.show()

else:
    print("Columns 'ROI' or 'Channel_Used' not found in the dataframe.")


# Tabulate 6 boxplots (if you have 6 or fewer unique channels)

unique_channels = df['Channel_Used'].unique()
num_channels = len(unique_channels)

if num_channels <= 6:
  plt.figure(figsize=(15, 8))  # Adjust figure size as needed

  for i, channel in enumerate(unique_channels):
    plt.subplot(2, 3, i + 1)  # Create subplots in a 2x3 grid
    channel_data = df[df['Channel_Used'] == channel]
    sns.boxplot(y='ROI', data=channel_data)
    plt.title(f'ROI for {channel}')
    plt.ylabel('')  # Remove y-axis label for cleaner look

  plt.tight_layout()  # Adjust subplot parameters for a tight layout
  plt.show()

elif num_channels > 6:
    print("You have more than 6 unique channels. Consider adjusting the visualization or channel grouping.")
```
